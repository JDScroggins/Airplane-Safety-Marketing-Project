```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dogs_df = pd.read_excel('hotdog-contest-winners-Copy1.xlsm')
```

```python
dogs_df.head()
```

```python
plt.rcParams['figure.figsize'] = (15,15) 
dogs_group = dogs_df.groupby('Country')["Dogs eaten"].sum()
dogs_group.plot(kind = 'bar')
plt.xlabel('Country', size = 15)
plt.ylabel('Hot Dogs Eaten', size = 15)
plt.title('Hot Dogs Eaten By Country', size = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (15,15) 
dogs_df.groupby(['Country','Winner']).sum().unstack().plot(kind='bar',y='Dogs eaten', stacked=True)
plt.xlabel('Country', size = 15)
plt.ylabel('Hot Dogs Eaten', size = 15)
plt.title('Hot Dogs Eaten By Country per Winner', size = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,10) 
dogs_group = dogs_df.groupby('Country')["Dogs eaten"].sum()
dogs_group.plot(kind = 'pie',  autopct='%1.1f%%')
plt.title('Hot Dogs Eaten By Country', size = 20)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,10) 
dogs_group = dogs_df.groupby('Country')["Dogs eaten"].sum()
dogs_group.plot(kind = 'pie',  autopct='%1.1f%%')
plt.title('Hot Dogs Eaten By Country', size = 20)
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)
plt.show()
```

```python
plt.rcParams['figure.figsize'] = (10,10) 
dogs_df.plot(x = 'Year', y = 'Dogs eaten')
plt.xlabel('Year', size = 15)
plt.ylabel('Hot Dogs Eaten', size = 15)
plt.title('Hot Dogs Eaten per Year', size = 20)
plt.show()
```
